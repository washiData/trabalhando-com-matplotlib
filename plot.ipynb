```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('C:\gas.csv')
gas
```

```python
plt.plot(gas['Year'], gas['Australia']) # plotando coluna em especifico
```

```python
plt.title('Valor do Combustivel em US$')# adicionando titulo

plt.plot(gas['Year'], gas['Australia'],'r.-', label = 'Australia')# label adicionado legenda ao grafico em especfico
plt.plot(gas['Year'], gas['Italy'],'y.-', label = 'Italy')
plt.plot(gas['Year'],gas['USA'],'b.-', label = 'USA')
plt.plot(gas['Year'],gas['Japan'],'g.-', label = 'Japan')

plt.xticks(gas['Year'] [:: 2])#[:: 2]) isso siginifica do inicio ao fim em dois em dois anos 

plt.legend()# colocando legenda no grafico
plt.show()
```

```python
plt.title('Price is US$') # colocando titulo no grafico

list_country = ['Australia', 'USA', 'Italy', 'Mexico', 'France', 'Germany'] # usando um loop

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label=[country], marker='.') # marker colocando ponto nas linhas do grafico
        
        
plt.xticks(gas['Year'] [:: 2])#[:: 2]) isso siginifica do inicio ao fim em dois em dois anos 

plt.xlabel('Year') # colocando legenda nos eixos do grafico x e y
plt.ylabel('US$')# colocando legenda nos eixos do grafico x e y
plt.savefig('country.pdf') # salvando o grafico em pdf e png
plt.savefig('country.png') # salve sempre antes do show
plt.legend()
plt.show()        
```
